```python
import pickle
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm  # For finding the Times New Roman font
```

### dcm eval

```python
#desiderate: [label value, object value, positional information]
content_llama =   [0, 0.03, 0.13]
duplicate_llama = [0, 0.02, 0.49]
position_llama =  [0, 0.39, 0.1]
value_llama =     [0, 0.07, 0.43]

content_goat =    [0, 0.02, 0.18]
duplicate_goat =  [0, 0.02, 0.55]
position_goat =   [0, 0.54, 0.15] 
value_goat =      [0, 0.06, 0.45]
```

```python
# set width ofbar
barWidth = 0.3
num_figs = 4

llama = [content_llama, duplicate_llama, position_llama, value_llama]
goat =  [content_goat,  duplicate_goat,  position_goat,  value_goat]
desiderata = ['label value', 'object value', 'position']
groups = ['Contet Gather', 'Duplicate Token', 'Positional Info Fetcher', 'Object Value Fetcher']
fig, axs = plt.subplots(1,num_figs,figsize=(15, 2))
plt.subplots_adjust(hspace=0.05,wspace=0.05)

font_properties = {'family': 'FreeMonoBoldOblique', 'size': 12}

for i in range(len(groups)):
    llama_val = llama[i]
    goat_val = goat[i]
    
    # Set position of bar on X axis
    br1 = np.arange(len(llama_val))
    br2 = [x + barWidth for x in br1]

    # Make the plot
    axs[i].bar(br1, llama_val, color ='grey', alpha=0.5, width = barWidth,
            label ='Llama-7b')
    axs[i].bar(br2, goat_val, color ='orange', alpha=0.5, width = barWidth,
            label ='Goat-7b')

    # Adding Xticks
    axs[i].set_xticks([r + barWidth for r in br1])  # Set the tick positions
    axs[i].set_xticklabels(desiderata, fontsize = 12)  # Set the tick labels
    axs[i].tick_params(axis="x", labelrotation=15) 
#     axs[i].grid(axis='y')
    axs[i].set_ylim(0,0.6)
    axs[i].set_title(groups[i])


    if i==0:
        axs[i].set_ylabel('Accuracy', fontsize = 12)
#     if i==3:    
        axs[i].legend(fontsize = 12, ncol=1, loc='upper left')
        axs[i].grid(True, axis='y', linestyle='--', alpha=0.7)

    else:
        axs[i].grid(True, axis='y', linestyle='--', alpha=0.7)
#         axs[i].set_yticks([    ])
        tick_positions = axs[i].get_yticks()
        tick_labels = axs[i].get_yticklabels()

        # Set the tick positions and empty labels
        axs[i].set_yticks(tick_positions)
        axs[i].set_yticklabels([''] * len(tick_labels))  # Set empty labels


```

### CMAP

```python
head_gropus = ['Contet Gather', 'Duplicate Token', 'Positional Info', 'Object Value', 'Positional Info +\nObject Value', 'Full Circuit']
cmap_heads = [0.68, 0.66, 0.78, 0.76, 0.79, 0.79]
llama_full = 0.656
goat_full = 0.822
inputs_pos = ['Final Query', 'Keys', 'Values','Keys, Final Query', 'Keys, Final Query,\nLable position val.', 'Keys, Final Query,\nAll values'] 
cmap_pos = [0.6967, 0.6167, 0.7367, 0.7133, 0.7467, 0.7833]
inputs_val = ['Final Query', 'Keys', 'Values','Keys, Final Query', 'Keys, Final Query,\nLable position val.', 'Keys, Final Query,\nAll values'] 
cmap_val = [0.7517, 0.6667, 0.69, 0.7533, 0.76, 0.765]
```

```python
titles = ['Circuit components', 'Positional Info Fetcher Heads', 'Object Value Fetcher Heads']
values = [cmap_heads, cmap_pos, cmap_val]
barWidth = 0.5
num_figs = len(titles)
labels = [head_gropus,inputs_pos,inputs_val]
fig, axs = plt.subplots(1,num_figs,figsize=(15, 2))
plt.subplots_adjust(hspace=0.05,wspace=0.05)

for i in range(num_figs):
    val = values[i]
    val
    br1 = np.arange(len(val))
    # Make the plot
    axs[i].plot(range(-1,len(val)+1), llama_full*np.ones(len(val)+2), '--', alpha=1, label='Llama-7b (full model)')
    axs[i].plot(range(-1,len(val)+1), goat_full*np.ones(len(val)+2), '--', color='orange', alpha=1, label='Goat-7b (full model)')
    axs[i].bar(br1, val, color ='grey', alpha=0.5, width = barWidth, label =titles[i])
    

    # Adding Xticks
    axs[i].set_xticks(br1)  # Set the tick positions
    axs[i].set_xticklabels(labels[i], fontsize = 11)  # Set the tick labels
    axs[i].tick_params(axis="x", labelrotation=45) 
    axs[i].set_xlim(-0.5,5.5)
    axs[i].set_ylim(0.5,0.85)
    axs[i].set_title(titles[i])


    if i==0:
        axs[i].set_ylim(0.5,0.85)
        axs[i].set_ylabel('Accuracy', fontsize = 11)
        axs[i].grid(True, axis='y', linestyle='--', alpha=0.7)
#     if i==3:    
#         axs[i].legend(fontsize = 12, ncol=2, frameon=False, loc='upper left')
        axs[i].legend(fontsize = 11,handles=[axs[i].lines[0],axs[i].lines[1]], labels=['Llama-7b (full model)','Goat-7b (full model)'])
    else:
        axs[i].set_ylim(0.5,0.85)
        axs[i].grid(True, axis='y', linestyle='--', alpha=0.7)
#         tick_positions = axs[i].get_yticks()
        tick_labels = axs[i].get_yticklabels()

        # Set the tick positions and empty labels
#         axs[i].set_yticks(tick_positions)
        axs[i].set_yticklabels([''] * len(tick_labels))  # Set empty labels

```
